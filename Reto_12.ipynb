2. Desarrollar una funcion que reciba dos diccionarios como parametros y los mezcle, es decir, que se construya un nuevo diccionario con las llaves de los dos diccionarios; si hay una clave repetida en ambos diccionarios, se debe asignar el valor que tenga la clave en el primer diccionario.

```python
def mezcla_diccionarios(dict_1, dict_2):
    mezcla_dict = dict_2.copy() # Copiamos el segundo diccionario
    mezcla_dict.update(dict_1) # Mencionamos el primer diccionario para actualizar si se repiten los valores
    return mezcla_dict

# Caso con dos diccionarios
dict_1 = {20:"dinero", 44:"sudor", 33:"trabajar"}
dict_2 = {33: "ricos", 22: "facilidad", 67:"relajación"}

resultado = mezcla_diccionarios(dict_1, dict_2)
print(resultado)
```

```python
import json  # Importamos la librería JSON para manejar archivos en este formato

def cargar_datos(archivo):
    """Carga los datos del archivo JSON y los devuelve como un diccionario."""
    with open(archivo, "r", encoding="utf-8") as file:  # Abrimos el archivo en modo lectura
        return json.load(file)  # Cargamos y devolvemos el contenido en formato diccionario

def buscar_por_deporte(datos, deporte):
    """Devuelve una lista con los nombres completos de las personas que practican el deporte ingresado."""
    resultado = []  # Lista para almacenar los nombres de las personas que coincidan
    for usuario in datos.values():  # Recorremos cada usuario dentro del diccionario
        if deporte in usuario["deportes"]:  # Verificamos si el deporte está en la lista de deportes del usuario
            resultado.append(f'{usuario["nombres"]} {usuario["apellidos"]}')  # Guardamos el nombre completo
    return resultado  # Retornamos la lista con los nombres encontrados

def buscar_por_edad(datos, edad_min, edad_max):
    """Devuelve una lista con los nombres completos de las personas dentro del rango de edad dado."""
    resultado = []  # Lista para almacenar los nombres de las personas dentro del rango de edad
    for usuario in datos.values():  # Recorremos cada usuario dentro del diccionario
        if edad_min <= usuario["edad"] <= edad_max:  # Verificamos si la edad está dentro del rango
            resultado.append(f'{usuario["nombres"]} {usuario["apellidos"]}')  # Guardamos el nombre completo
    return resultado  # Retornamos la lista con los nombres encontrados

# Nombre del archivo JSON (debe estar en la misma carpeta que este script)
archivo_json = "personas.json"

# Cargamos los datos del JSON en un diccionario
datos = cargar_datos(archivo_json)

# Solicitar deporte al usuario
deporte_ingresado = input("Ingrese un deporte: ")  # Pedimos al usuario que ingrese un deporte
personas_deporte = buscar_por_deporte(datos, deporte_ingresado)  # Buscamos las personas que lo practican

# Imprimimos el resultado
if personas_deporte:
    print("\nPersonas que practican", deporte_ingresado, ":", personas_deporte)
else:
    print("\nNo hay personas que practiquen ese deporte.")

# Solicitar rango de edad al usuario
edad_min = int(input("Ingrese la edad mínima: "))  # Pedimos la edad mínima
edad_max = int(input("Ingrese la edad máxima: "))  # Pedimos la edad máxima

# Buscamos las personas dentro del rango de edad
personas_edad = buscar_por_edad(datos, edad_min, edad_max)

# Imprimimos el resultado
if personas_edad:
    print("\nPersonas en el rango de edad:", personas_edad)
else:
    print("\nNo hay personas dentro de ese rango de edad.")
```

5. A través de un programa conectese a al menos 3 API's , obtenga el JSON, imprimalo y extraiga los pares de llave : valor.

```python
import requests
import json

# Definir las API a consultar
apis = {
    "Cat Facts": "https://catfact.ninja/fact",
    "Dog API": "https://dog.ceo/api/breeds/image/random",
    "Public Holidays": "https://date.nager.at/api/v3/PublicHolidays/2024/US"
}

# Función para hacer la petición y extraer los datos
def fetch_api_data(api_name, url):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Lanza un error si la respuesta no es 200 OK
        data = response.json()

        print(f"\nDatos obtenidos de {api_name}:\n")
        print(json.dumps(data, indent=4))  # Imprimir JSON formateado

        print("\nPares clave-valor extraídos:")
        for key, value in data.items() if isinstance(data, dict) else enumerate(data):
            print(f"{key}: {value}")

    except requests.exceptions.RequestException as e:
        print(f"Error al obtener datos de {api_name}: {e}")

# Recorrer y obtener datos de cada API
for name, url in apis.items():
    fetch_api_data(name, url)
```
